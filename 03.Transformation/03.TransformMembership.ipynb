#### QUERY AND VALIDATE MEMBERSHIPS FROM `GIZMO.BRONZE.MEMBERSHIPS_VW`

```sql
SELECT * FROM gizmo.bronze.memberships_vw
```

#### EXTRACT CUSTOMER_ID FROM PATH 

```sql
SELECT
  regexp_extract(path, '.*/([0-9]+)\\.png$', 1) AS customer_id,
  content as membership_card
FROM gizmo.bronze.memberships_vw;
```

#### CREATE `GIZMO.SILVER.MEMBERSHIPS`
1. CATALOG NAME: GIZMO
2. SCHEMA NAME: SILVER
3. TABLE NAME: MEMBERSHIPS

```sql
CREATE OR REPLACE TABLE gizmo.silver.memberships
AS
SELECT
  regexp_extract(path, '.*/([0-9]+)\\.png$', 1) AS customer_id,
  content as membership_card
FROM gizmo.bronze.memberships_vw;
```
