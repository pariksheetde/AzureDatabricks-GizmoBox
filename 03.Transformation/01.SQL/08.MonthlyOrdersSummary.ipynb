#### MONTHLY ORDER SUMMARY
1. TOTAL ORDERS
2. TOTAL ITEMS BOUGHT
3. TOTAL AMOUNT SPENT

#### WRITE TRANSFORMED MONTHLY ORDER SUMMARY DATA TO GOLD SCHEMA
1. CATALOG NAME: GIZMO
2. SCHEMA NAME: GOLD
3. TABLE NAME: MONTHLY_ORDERS_SUMMARY

```sql
CREATE OR REPLACE TABLE GIZMO.GOLD.MONTHLY_ORDERS_SUMMARY
AS
SELECT
date_format(transaction_timestamp, 'yyyy-MM') AS order_month,
customer_id,
COUNT(distinct order_id) AS total_orders,
SUM(quantity) AS total_item_bought,
SUM(price * quantity) AS total_amount
FROM
GIZMO.SILVER.ORDERS
GROUP BY customer_id, date_format(transaction_timestamp, 'yyyy-MM');
```
