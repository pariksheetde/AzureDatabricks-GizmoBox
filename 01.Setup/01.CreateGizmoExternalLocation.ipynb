### SET-UP PROJECT ENVIRONMENT FOR `GIZMO` DATA LAKEHOUSE

#### CREATE `STORAGE LOCATION` FOR `GIZMO` CONTAINER
1. CONTAINER NAME: gizmo
2. ACCOUNT NAME: gizmoboxdevadls

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS gizmo_external_location
URL 'abfss://gizmo@gizmoboxdevadls.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `gizmobox-dev-storage-credential`)
COMMENT 'CREATING EXTERNAL LOCATION FOR GIZMO CONTAINER'
```

```sql
SHOW CATALOGS;
```

#### CREATE `CATALOG` FOR `GIZMO` CONTAINER
1. CATALOG NAME: gizmo
2. CONTAINER NAME: gizmoboxdevadls

```sql
CREATE CATALOG IF NOT EXISTS gizmo
MANAGED LOCATION 'abfss://gizmo@gizmoboxdevadls.dfs.core.windows.net/'
COMMENT 'CREATING CATALOG FOR GIZMO';
```

#### CREATE SCHEMAS
1. Landing
2. Bronze
3. Silver
4. Gold

```sql
USE CATALOG gizmo;
SELECT current_catalog();
```

#### CREATE `LANDING` SCHEMA

```sql
CREATE SCHEMA IF NOT EXISTS landing
COMMENT 'CREATE LANDING SCHEMA'
MANAGED LOCATION 'abfss://gizmo@gizmoboxdevadls.dfs.core.windows.net/landing';
```

#### CREATE `BRONZE` SCHEMA

```sql
CREATE SCHEMA IF NOT EXISTS bronze
COMMENT 'CREATE BRONZE SCHEMA'
MANAGED LOCATION 'abfss://gizmo@gizmoboxdevadls.dfs.core.windows.net/bronze';
```

#### CREATE `SILVER` SCHEMA

```sql
CREATE SCHEMA IF NOT EXISTS silver
COMMENT 'CREATE SILVER SCHEMA'
MANAGED LOCATION 'abfss://gizmo@gizmoboxdevadls.dfs.core.windows.net/silver';
```

#### CREATE `GOLD` SCHEMA

```sql
CREATE SCHEMA IF NOT EXISTS gold
COMMENT 'CREATE GOLD SCHEMA'
MANAGED LOCATION 'abfss://gizmo@gizmoboxdevadls.dfs.core.windows.net/gold';
```

```sql
SHOW SCHEMAS;
```

#### CREATE VOLUME

```sql
USE CATALOG gizmo;
USE SCHEMA landing;
CREATE EXTERNAL VOLUME IF NOT EXISTS operational_data
LOCATION 'abfss://gizmo@gizmoboxdevadls.dfs.core.windows.net/landing/operational'
COMMENT 'CREATE VOLUME opearation_data';
```
