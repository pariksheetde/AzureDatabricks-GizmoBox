#### CREATE `PAYMENTS_EXTERNAL`
- CATALOG NAME: gizmo
- SCHEMA NAME: bronze
- EXTERNAL TABLE: payments_external

```sql
CREATE TABLE IF NOT EXISTS gizmo.bronze.payments_external
(
  payment_id INTEGER,
  customer_id INTEGER,
  payment_date TIMESTAMP,
  payment_status INTEGER,
  payment_method STRING
)
USING CSV
OPTIONS
(
  header = "true",
  delimiter = ","
)
LOCATION 'abfss://gizmo@gizmoboxdevadls.dfs.core.windows.net/landing/external/payments'
```

```sql
REFRESH TABLE gizmo.bronze.payments_external
```
