#### READ `MEMBERSHIPS` FROM VOLUME IN SPARK DATAFRAME
1. EXTRACT FILE_NAME AS METADATA
2. ADD `LOAD_TIMESTAMP` AS AUDIT COLUMN

```python
membership_df = spark.read.format("binaryFile").load(
    '/Volumes/gizmo/landing/operational_data/memberships/*/*.png'
)
display(membership_df.limit(6))
```

#### WRITE `MEMBERSHIPS` INTO `DELTA FORMAT`

```python
membership_df.writeTo('gizmo.bronze.memberships_delta').createOrReplace()
```

#### VALIDATE AND QUERY THE TABLE `GIZMO.BRONZE.MEMBERSHIPS_DELTA`

```python
%sql
SELECT * FROM GIZMO.BRONZE.MEMBERSHIPS_DELTA;
```
