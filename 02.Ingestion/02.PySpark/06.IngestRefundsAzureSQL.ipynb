### QUERY FROM AZURE SQL
1. DATABASE: GizmoDB
2. SERVER: gizmosrvr

#### PYSPARK DATAFRAME TO READ FROM `AZURE SQL via JDBC`

```python
refunds_df = spark.read.format("jdbc").options(
    url="jdbc:sqlserver://gizmosrvr.database.windows.net:1433;database=GizmoDB",
    dbtable="refunds",
    user="admins",
    password="Gubi@123"
).load()
```

#### WRITE THE DATAFRAME TO `HIVE METASTORE`

```python
refunds_df.writeTo('gizmo.bronze.refunds_delta').createOrReplace()
```
