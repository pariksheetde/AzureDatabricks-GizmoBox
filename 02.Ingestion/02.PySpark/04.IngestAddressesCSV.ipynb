#### READ `ADDRESSES` FROM VOLUME IN SPARK DATAFRAME
1. EXTRACT FILE_NAME AS METADATA
2. ADD `LOAD_TIMESTAMP` AS AUDIT COLUMN

#### READ USING SPARK

```python
from pyspark.sql.functions import col, current_timestamp

address_df = (spark.read.format("csv")
              .option("header", "true")
              .option("delimiter", "\t")
              .load("/Volumes/gizmo/landing/operational_data/addresses/")
              .withColumn("load_timestamp", current_timestamp()))
display(address_df)
```

#### WRITE `ADDRESSES` INTO `DELTA FORMAT`

```python
address_df.writeTo("gizmo.bronze.addresses_delta").createOrReplace()
```

#### VALIDATE AND QUERY THE TABLE GIZMO.BRONZE.ADDRESSES_DELTA

```python
%sql
SELECT * FROM GIZMO.BRONZE.ADDRESSES_DELTA ORDER BY 1;
```
