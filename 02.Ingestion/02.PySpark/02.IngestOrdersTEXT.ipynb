#### READ `ORDERS` FROM VOLUME IN SPARK DATAFRAME
1. EXTRACT FILE_NAME AS METADATA
2. ADD `LOAD_TIMESTAMP` AS AUDIT COLUMN

```python
from pyspark.sql.functions import current_timestamp

orders_df = (spark.read.text('/Volumes/gizmo/landing/operational_data/orders/')
                   .withColumn("load_timestamp", current_timestamp())
)
display(orders_df)
```

#### WRITE `ORDERS` INTO `DELTA FORMAT`

```python
orders_df.writeTo("gizmo.bronze.orders_delta").createOrReplace()
```

#### VALIDATE AND QUERY THE TABLE `GIZMO.BRONZE.ORDERS_DELTA`

```python
%sql
SELECT * FROM GIZMO.BRONZE.ORDERS_DELTA;
```
