#### CREATE `ADDRESSES_VW`
- CATALOG NAME: gizmo
- SCHEMA NAME: bronze
- VIEW NAME: addresses_vw

```python
%sql
CREATE OR REPLACE VIEW gizmo.bronze.addresses_vw
AS
SELECT * FROM read_files('/Volumes/gizmo/landing/operational_data/addresses/', 
  FORMAT => 'csv',
  delimiter => '\t', 
  header => 'true')
```
